```python
%matplotlib inline
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Input, Dense
from keras.models import Model
import numpy as np
```

```python
def read_images(dir):
    # rescales images
    datagen = ImageDataGenerator(rescale=1./255)
    
    generator = datagen.flow_from_directory(
        dir,
        target_size = (150,150),
        batch_size = 20,
        class_mode = 'input',
        color_mode = 'rgb'
    )
    
    return(generator)
```

```python
datasets = {
    'train': 'data/training', 
    'test': 'data/testing', 
    'valid': 'data/validation',}

key_to_generator = {k: read_images(v) for k,v in datasets.items()}
```

```python
# dimensions of our images.
img_width, img_height = 150, 150
input_img = Input(shape=(22500,))
input_img
```

```python
# this is the size of our encoded representations
encoding_dim = 50  

# "encoded" is the encoded representation of the input
encoded = Dense(encoding_dim, activation='relu')(input_img)
# "decoded" is the lossy reconstruction of the input
decoded = Dense(22500, activation='sigmoid')(encoded)

# this model maps an input to its reconstruction
autoencoder = Model(input_img, decoded)

# this model maps an input to its encoded representation
encoder = Model(input_img, encoded)

# create a placeholder for an encoded (32-dimensional) input
encoded_input = Input(shape=(encoding_dim,))
# retrieve the last layer of the autoencoder model
decoder_layer = autoencoder.layers[-1]
# create the decoder model
decoder = Model(encoded_input, decoder_layer(encoded_input))

autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
autoencoder.fit_generator(key_to_generator['train'],
                epochs=50,
                shuffle=True,
                validation_data=key_to_generator['test']
                )
```
